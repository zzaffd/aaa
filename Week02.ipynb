### 2.2.5 DataFrame - index, columns  변경하기
* 행인덱스 변경: **<u>DataFrame객체.index = [새인덱스리스트]</u>**
* 열 이름 변경: **<u>DataFrame객체.columns = [새컬럼명리스트]</u>**
* 행인덱스 변경: **<u>DataFrame객체.rename(index={기존인덱스명:새인덱스명,...})</u>**
* 열 이름 변경: **<u>DataFrame객체.rename(columns={기존컬럼명:새컬럼명,...})</u>**

  
* 특정 열을 행 인덱스로 설정 : **<u>DataFrame객체.set_index(['열이름'] 또는 '열이름')</u>**
* 행인덱스 초기화 : **<u>DataFrame객체.reset_index()</u>**
    * 행 인덱스를 정수형 위치 인덱스로 초기화한다. 이때 기존 행 인덱스는 열로 이동한다.

```python
# df 확인

df
```

```python
# df의 인덱스를 '음료1', '음료2', '음료3'으로 바꾸기

df.index=['음료1','음료2','음료3']
```

```python
# df의 컬럼명을 '음료명', '판매가격', '판매수량'으로 바꾸기

df.columns=['음료명','판매가격','판매수량']
```

```python
# df의 '음료명'컬럼명을 '제품명'으로 바꾸고, '음료1'인덱스를 'coffee'로 바꾸기

df.rename(columns={'음료명':'제품명'}, index={'음료1':'coffee'})
```

```python
# df확인 , 변경 안됨

df
```

df=df.rename(columns={'음료명':'제품명'}, index={'음료1':'coffee'}) 

이렇게도 적용가능.

```python
# inplace=True

df.rename(columns={'음료명':'제품명'}, index={'음료1':'coffee'}, inplace=True)
```

```python
# df확인, 변경됨

df
```

```python
# '제품명' 컬럼을 인덱스로 설정한다.

df.set_index('제품명')
```

```python
# df확인, 변경안됨

df
```

```python
# df.set_index('제품명', inplace=True)와 동일한 효과를 내기위해
# df.set_index()한 결과를 df변수에 대입

df.set_index('제품명', inplace=True)
```

```python
# df확인

df
```

```python
# df의 인덱스를 정수형 위치 인덱스로 초기화한다.
# 결과를 df변수에 대입

df.reset_index()
```
